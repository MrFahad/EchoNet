```python
import os, random
from moviepy import VideoFileClip

from moviepy.editor import VideoFileClip

clip = VideoFileClip("input.mkv", audio=True)


INPUT_FOLDER = "DATA/Laliga"
OUTPUT_FOLDER = "DATA/Laliga_Out"
FRAMES_PER_CLIP = 500

os.makedirs(OUTPUT_FOLDER, exist_ok=True)

def process_video(in_path, out_path, n_frames):
    clip = VideoFileClip(in_path)
    fps = clip.fps
    total_frames = int(clip.duration * fps)

    if total_frames < n_frames:
        print(f"⚠️ {os.path.basename(in_path)} shorter than {n_frames} frames — skipping.")
        clip.close()
        return

    start_frame = random.randint(0, total_frames - n_frames)
    t0 = start_frame / fps
    t1 = (start_frame + n_frames) / fps

    sub = clip.subclipped(t0, t1)
    sub.write_videofile(
        out_path,
        codec="libx264",
        audio_codec="aac",
        logger=None
    )

    print(f"✅ Saved {out_path} | frames {start_frame}–{start_frame + n_frames}")
    clip.close()
    sub.close()

def main():
    for fname in os.listdir(INPUT_FOLDER):
        if fname.lower().endswith((".mp4", ".mkv", ".avi", ".mov")):
            inp = os.path.join(INPUT_FOLDER, fname)
            base, ext = os.path.splitext(fname)
            outp = os.path.join(OUTPUT_FOLDER, f"{base}_out{ext}")
            process_video(inp, outp, FRAMES_PER_CLIP)

if __name__ == "__main__":
    main()
```
